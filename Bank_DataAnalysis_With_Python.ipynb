```python
##                             Phase 2 – Transaction Analysis
```

```python
##Transection Analysis
#We will compute:
    
#Total transactions
#Total amount by transaction_type
#Account_type wise amount
#City wise volume
#Monthly trends
#Failed transaction %
```

```python
#Total transactions
df['transaction_id'].count()
```

```python
##Total amount by transaction_type
sales_by_transection_type = df.groupby('transaction_type')['amount'].sum()
sales_by_transection_type
```

```python
##Account_type wise amount
Account_Type_Amount = df.groupby('account_type')['amount'].sum()
Account_Type_Amount
```

```python
##City wise volume
City_Wise_Volume = df.groupby('city')['transaction_id'].count().sort_values(ascending=False)
City_Wise_Volume
```

```python
##Monthly trends
df['month_year'] = df['transaction_date'].dt.to_period('M')

Monthly_Trends = df.groupby('month_year')['amount'].sum()
Monthly_Trends 
```

```python
##Failed transaction %
Faild_Trans = (df[df['status'] == 'Failed'].groupby('status')['status'].count() * 100) / df['status'].count() 
Faild_Trans
```

```python
##                             Phase 3 – Customer & Channel Insights
```

```python
##Customer & Channel Insights
#Top 5 customers by value

#Low / negative balance customers

#Channel contribution %

#Avg transaction per channel

#High-withdrawal anomaly detection
```

```python
#Top 5 customers by value
Top_5_Customer_BY_Value = df.groupby(['customer_id','customer_name'])['amount'].sum().sort_values(ascending=False).head(5)
Top_5_Customer_BY_Value
```

```python
#Low / negative balance customers
df[df['balance'] < 1000]
```

```python
#Channel contribution %
Channel_Contribution = round((df.groupby('channel')['amount'].sum()/df['amount'].sum()).sort_values(ascending=False) * 100)
Channel_Contribution
```

```python
#Avg transaction per channel
Avg_Transection_Per_Channel = df.groupby('channel')['amount'].mean()
Avg_Transection_Per_Channel
```

```python
##High-withdrawal anomaly detection
High_Withdrawls = df[df['transaction_type']=='Withdrawal'].groupby('customer_id')['amount'].sum().sort_values(ascending=False).head(5)
High_Withdrawls 
```
